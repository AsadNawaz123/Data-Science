## Question 1 - Extracting Tesla Stock Data Using yfinance

```python
import yfinance as yf

# Fetch Tesla stock data
df = yf.download('TSLA', start='2023-01-01', end='2023-12-31')
df.head()
```

## Question 2 - Extracting Tesla Revenue Data Using Webscraping

```python
from bs4 import BeautifulSoup
import requests

#URL for scraping Tesla's revenue data
url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'

# Send a GET request
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find all elements containing revenue data
revenues = soup.find_all('div', class_='revenue-data')

# Check the length of revenues list before accessing specific indexes
if len(revenues) > 5:
    print(revenues[5])  # Access the element at index 5
else:
    print("Insufficient elements in 'revenues' list or index out of range.")
df.tail()
```

## Question 5 - Tesla Stock and Revenue Dashboard

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have the Tesla stock data in a DataFrame named df

def make_graph(data, title):
    plt.figure(figsize=(10, 6))  # Set the size of the figure
    plt.plot(data.index, data['Close'], label='Tesla Stock Close Price', color='blue')  # Plotting the 'Close' prices
    plt.title(title)  # Set the title for the plot
    plt.xlabel('Date')  # Label for x-axis
    plt.ylabel('Stock Price')  # Label for y-axis
    plt.legend()  # Show legend
    plt.grid(True)  # Show grid
    plt.show()  # Display the plot

# Use the make_graph function to plot Tesla stock data with a specific title
make_graph(df, title='Tesla Stock Price (2023)')
```
